```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
#(a)Generate 50 data points from this function in the range [â€“ 3, 3]. 
np.random.seed(50)
X = np.random.default_rng().uniform(-3,3,50) 
y =(np.sin(np.pi*X))/(np.pi*X)
plt.scatter(X,y)
plt.show()

#(b)Add Gaussian noise to the data.
m = 3
s = 3
gnoise = np.random.normal(m, s, 50) 
y_gnoise=(np.sin(np.pi*X))/(np.pi*X)+gnoise
plt.scatter(X,y_gnoise)
plt.show()
```

```python
#(c)Train an SVM regressor with the data generated in (a). Define (and explain) suitable parameters required 
# for training the regressor. 
X = X.reshape(-1,1)
y = y.reshape(-1,1)

sc = StandardScaler()
X = sc.fit_transform(X)
y = sc.fit_transform(y)
y = y.reshape(50,)

Reg = SVR(kernel = 'rbf')
Reg.fit(X,y)

# Visualization.
plt.scatter(X,y)
plt.scatter(X, Reg.predict(X), color = 'black')
plt.title('Radial Basis Func.')
plt.show()
```
